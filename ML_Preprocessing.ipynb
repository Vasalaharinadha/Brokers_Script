```python
import dask.dataframe as dd

# Load your datasets as Dask DataFrames
angel = dd.from_pandas(angel, npartitions=4)
fyers = dd.from_pandas(fyers, npartitions=4)
kotak = dd.from_pandas(kotak, npartitions=4)
icic = dd.from_pandas(icic, npartitions=4)
zeroba = dd.from_pandas(Zeroba, npartitions=4)

# Merge using Dask
merged_data = angel.merge(fyers, on='Exch', how='outer').merge(kotak, on='Exch', how='outer').merge(icic, on='Exch', how='outer').merge(zeroba, on='Exch', how='outer')

# Compute the result
result = merged_data.compute()

# Display the merged DataFrame
result.head()
```

```python
#When we are use SQl this is the Query

# here i take Exch column and mapping all dataset

SELECT *
FROM angel AS a
INNER JOIN icic AS i ON a.Exch = i.Exch
INNER JOIN kotak AS k ON a.Exch = k.exchange
INNER JOIN fyers AS f ON a.Exch = f.Exch
INNER JOIN zeroba AS z ON a.Exch = z.Exch;
```

```python
# here i take TradingSymbol column and mapping all dataset

SELECT *
FROM angel AS a
INNER JOIN icic AS i ON a.Exch = i.Exch AND a.TradingSymbol = i.TradingSymbol
INNER JOIN kotak AS k ON a.Exch = k.exchange AND a.Token = k.instrumentToken
INNER JOIN fyers AS f ON a.Exch = f.Exch AND a.TradingSymbol = f.TradingSymbol
INNER JOIN zeroba AS z ON a.Exch = z.Exch AND a.TradingSymbol = z.TradingSymbol;
```
